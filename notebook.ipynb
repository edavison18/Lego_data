# "Only the Best is Good Enough"

The goal of this project is to explore the lego data from 1970 to 2022 to identify the evolution of Lego as a company creating products for young children to adults. 

Questions to guide this exploration: 

1. What are the largest Themes/theme groups?
2. What are the most expensive/least expensive sets and do those sets get more expensive over time?
3. Average prices and peices by year, theme groups


First, I want to explore the data and identify the different themeGroups and themes.

```python
--Explore data and columns
Select * 
From lego_sets.csv as lego
```

```python
--  Count by ThemeGroup
Select themeGroup, Count(themeGroup) as total
From lego_sets.csv
Group by themeGroup
Order by Count(themeGroup) DESC
```

```python
--Count by Theme 
Select theme, Count(theme) as total
From lego_sets.csv
Group by theme
Order by Count(theme) DESC
```

The Miscellaneous themeGroup was the largest so I drilled down to see what themes were considered miscellaneous. 

```python
--MISCELLANEOUS Theme group drill down
SELECT name, theme, subtheme
FROM lego_sets.csv
WHERE themeGroup = 'Miscellaneous'
GROUP BY name, theme, subtheme;
```

Then I wanted to look into the average prices, average number of pieces and average minimum age range. This I can use to compare and identfy the least or most expensive sets Lego makes. 

```python
-- Find average # of pieces for sets and average price
SELECT Round(AVG(pieces)) as avg_num_pieces, ROUND(AVG(US_retailPrice), 2) as avg_price, ROUND(AVG(agerange_min), 1) as avg_age_min 
FROM lego_sets.csv
```

```python
-- Top 10 Most expensive sets and year
Select name, year, pieces, max(US_retailPrice) as max_price 
From lego_sets.csv
WHERE US_retailPrice is not null
Group by name, year, pieces
Order by max(US_retailPrice) DESC
Limit 10

```

```python
-- Top 10 Least expensive sets and year
Select name, year, pieces, min (US_retailPrice) as min_price 
From lego_sets.csv
WHERE US_retailPrice is not null
Group by name, year, pieces
Order by min(US_retailPrice) ASC
Limit 10
```

```python
---Find average pieces and price group by year

SELECT year, Round(AVG(pieces)) as avg_num_pieces, ROUND(AVG(US_retailPrice), 2) as avg_price 
FROM lego_sets.csv
Group by year
Order by ROUND(AVG(US_retailPrice), 2) DESC
```

```python
---AVG price per different theme group (what sells)
SELECT themeGroup, ROUND(AVG(US_retailPrice), 2) as avg_price 
From lego_sets.csv
Group by themeGroup
Order by ROUND(AVG(US_retailPrice), 2) DESC
```

# Star Wars Over the Years

Questions:
1. Did Star Wars sets increase over the years?
2. Did the average number of pieces increase over the years? 
3. What was the average price each year and did that increase over the year? 
4. Group sets by expense to see how costly sets are.



```python
--Drill Down on Star Wars

SELECT name, year, pieces, minifigs, US_retailPrice
FROM lego_sets.csv
WHERE theme = 'Star Wars' and US_retailPrice is not null
Order by year DESC, US_retailPrice DESC
;
```

```python
--Star Wars Sets over the Years - Increase? Answer- Yes (drop after 2018 - last film 2019/covid??)

SELECT year, count(theme) as total_num_theme
FROM lego_sets.csv
WHERE theme = 'Star Wars'
Group by year
Order by year ASC
```

```python
--CREAT CTE FROM ABOVE table df9 AND Find avg price per year 

With star_wars_table AS (
SELECT name, year, pieces, minifigs, US_retailPrice
FROM lego_sets.csv
WHERE theme = 'Star Wars' and US_retailPrice is not null
Order by year DESC, US_retailPrice DESC
)

Select year, ROUND(AVG(pieces)) avg_num_pieces, Round(AVG(US_retailPrice), 2) as avg_price
From star_wars_table
Group by year
Order by year DESC, Round(AVG(US_retailPrice), 2)
```

```python
--- Dived prices into least expensive to most expensive categories

With expense_table AS (
Select  name, US_retailPrice, 
	CASE WHEN US_retailPrice > 100 THEN 'most expensive'
	WHEN US_retailPrice > 40 and US_retailPrice < 100 THEN 'slightly_expensive' 
	ELSE 'least_expensive' END as cost
FROM lego_sets.csv
WHERE theme = 'Star Wars' and US_retailPrice is not null)

SELECT cost, Count(cost) as num_sets_cost
From expense_table
Group by cost
Order by Count(cost) DESC
	
```
